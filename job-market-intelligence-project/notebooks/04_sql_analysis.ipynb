Remote vs onsite vs hybrid jobs


```python
query = """
SELECT
CASE
WHEN remote_ratio = 100 THEN 'Remote'
WHEN remote_ratio = 0 THEN 'Onsite'
ELSE 'Hybrid'
END AS work_mode,
COUNT(*) AS total_jobs
FROM job_market
GROUP BY work_mode
"""

df = pd.read_sql_query(query, conn)
df
```

Average salary by work mode

```python
query = """
SELECT
CASE
WHEN remote_ratio = 100 THEN 'Remote'
WHEN remote_ratio = 0 THEN 'Onsite'
ELSE 'Hybrid'
END AS work_mode,
AVG(salary_in_usd) AS avg_salary
FROM job_market
GROUP BY work_mode
ORDER BY avg_salary DESC
"""

df = pd.read_sql_query(query, conn)
df
```

Salary Statistics

```python
query = """
SELECT
MIN(salary_in_usd) AS min_salary,
MAX(salary_in_usd) AS max_salary,
AVG(salary_in_usd) AS avg_salary
FROM job_market
"""

df = pd.read_sql_query(query, conn)
df
```

Rank job titles by average salary
This query ranks job titles based on their average salary.

```python
query = """
SELECT
    job_title,
    AVG(salary_in_usd) AS avg_salary,
    RANK() OVER (ORDER BY AVG(salary_in_usd) DESC) AS salary_rank
FROM job_market
GROUP BY job_title
ORDER BY salary_rank
LIMIT 10
"""

df = pd.read_sql_query(query, conn)
df
```

Top Paying job in each Experience Level
This finds the highest paying job title inside each experience level.

```python
query = """
WITH ranked_jobs AS (
    SELECT
        experience_level,
        job_title,
        AVG(salary_in_usd) AS avg_salary,
        RANK() OVER (
            PARTITION BY experience_level
            ORDER BY AVG(salary_in_usd) DESC
        ) AS rank_within_experience
    FROM job_market
    GROUP BY experience_level, job_title
)
SELECT
    experience_level,
    job_title,
    avg_salary,
    rank_within_experience
FROM ranked_jobs
WHERE rank_within_experience = 1
ORDER BY avg_salary DESC
"""

df = pd.read_sql_query(query, conn)
df
```
