```python
jobs = pd.read_csv("C:/Users/riyak/OneDrive/Documents/job-market-intelligence-project/data/raw/adzuna_global_job_listings_2025.csv")
salary = pd.read_csv("C:/Users/riyak/OneDrive/Documents/job-market-intelligence-project/data/raw/DataScience_salaries_2024.csv")

jobs.head()
```

```python
jobs = jobs.drop_duplicates()
salary = salary.drop_duplicates()
```

```python
jobs.columns = jobs.columns.str.lower()
salary.columns = salary.columns.str.lower()
```

```python
jobs = jobs.drop(columns=['contract_time', 'contract_type'])
```

```python
jobs['company'] = jobs['company'].fillna('Unknown')
```

```python
jobs['salary_min'] = jobs['salary_min'].fillna(jobs['salary_min'].median())
jobs['salary_max'] = jobs['salary_max'].fillna(jobs['salary_max'].median())
```

```python
jobs = jobs.drop(columns=['latitude','longitude'])
```

```python
jobs.isnull().sum()
```

```python
jobs.head()
```

```python
jobs['title'] = jobs['title'].str.lower()
salary['job_title'] = salary['job_title'].str.lower()
```

```python
jobs = jobs.rename(columns={'title': 'job_title'})
jobs.columns
```
